```python
import pandas as pd
import os
import glob
```

#### This notebook combines all the field measurement files (for each gauge) into one big file to use in future analysis.
Field meaurement .csv files were obtained in field_measurements.ipynb notebook. Save database in field_measurements.csv.

```python
files = os.path.join("", "*.csv")
files = glob.glob(files)
files
```

```python
df = (pd.read_csv(file, skiprows = [1]) for file in files) #we skip 1st row of each file because it has invalid data from the original source
data = pd.concat(df, ignore_index=True)
```

```python
data
```

Documentation on the description of the columns and codes: https://help.waterdata.usgs.gov/output-formats#streamflow_measurement_data

```python
data.to_csv("field_measurements.csv")
```
