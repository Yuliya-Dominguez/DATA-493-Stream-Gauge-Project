### CSV Maker
This code will go in a loop to read each entry in caliornia_gages.csv and extract field measurements.
url -> to be replaced by an entry's *modified* url (I'll go over this later)
site_no -> to be replaced by an entry's site number

```python
for site_no in cal_gages["site_no"].values[1:]:
    url = "https://waterdata.usgs.gov/nwis/measurements?site_no=" + site_no + "&agency_cd=USGS&format=rdb_expanded"
    response = requests.get(url)
    if response.status_code == 200:
        print(site_no + " Success")
        data = pd.read_csv(url, sep='\t', lineterminator='\n', comment='#', header=0, skiprows=[29])
        data.to_csv("field_measurement_data/" + site_no + ".csv", index=False)
    else:
        print(site_no + " Error: ",response.status_code)
```
